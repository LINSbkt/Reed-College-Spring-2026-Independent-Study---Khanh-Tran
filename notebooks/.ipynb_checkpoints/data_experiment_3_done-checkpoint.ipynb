```python
import os
import json
import random
import numpy as np
import pandas as pd
import cv2
from PIL import Image
from segment_anything import sam_model_registry, SamPredictor

# ── PATHS ──────────────────────────────────────────────────────────────────
IMAGE_DIR   = 'C:/Users/Dell/Downloads/MIDOGpp/images/'
ANNO_FILE   = 'C:/Users/Dell/Downloads/MIDOGpp/databases/MIDOG++.json'
BASE_DIR    = 'C:/Users/Dell/Downloads/Independent-Study-Project---Mitotic-vqa/notebooks/patches_exp3'
SAM_CKPT    = 'C:/Users/Dell/Downloads/sam_vit_h.pth'

# ── SETTINGS ───────────────────────────────────────────────────────────────
CROP_SIZE   = 256
HALF        = CROP_SIZE // 2
RANDOM_SEED = 42

# Four off-center positions: (cell_x_in_crop, cell_y_in_crop)
# Cell will appear at this position within the 256x256 crop
POSITIONS = {
    0: (64,  64),   # top-left quadrant
    1: (192, 64),   # top-right quadrant
    2: (64,  192),  # bottom-left quadrant
    3: (192, 192),  # bottom-right quadrant
}

print("Setup complete")
print(f"Image directory : {IMAGE_DIR}")
print(f"Output directory: {BASE_DIR}")
print(f"Crop size       : {CROP_SIZE}x{CROP_SIZE}")
print(f"Positions       : {POSITIONS}")
```

```python
for condition in ['raw', 'bbox', 'mask']:
    for label in ['mitotic', 'hard_negative']:
        os.makedirs(f'{BASE_DIR}/{condition}/{label}', exist_ok=True)

print("Folder structure created:")
for condition in ['raw', 'bbox', 'mask']:
    for label in ['mitotic', 'hard_negative']:
        print(f"  patches_exp3/{condition}/{label}/")
```

```python
with open(ANNO_FILE, 'r') as f:
    data = json.load(f)

images_df      = pd.DataFrame(data['images'])
annotations_df = pd.DataFrame(data['annotations'])

# Only images we have downloaded (019 is missing)
downloaded_files = [
    '001.tiff', '002.tiff', '003.tiff', '004.tiff', '005.tiff',
    '006.tiff', '007.tiff', '008.tiff', '009.tiff', '010.tiff',
    '011.tiff', '012.tiff', '013.tiff', '014.tiff', '015.tiff',
    '016.tiff', '017.tiff', '018.tiff', '020.tiff', '022.tiff'
]

downloaded_images    = images_df[images_df['file_name'].isin(downloaded_files)]
downloaded_image_ids = downloaded_images['id'].tolist()
our_annotations      = annotations_df[annotations_df['image_id'].isin(downloaded_image_ids)]

print(f"Images available : {len(downloaded_images)}")
print(f"Total annotations: {len(our_annotations)}")
print(f"Mitotic          : {len(our_annotations[our_annotations['category_id']==1])}")
print(f"Hard negative    : {len(our_annotations[our_annotations['category_id']==2])}")
```

```python
random.seed(RANDOM_SEED)

# Assign a random position (0-3) to each annotation
position_assignments = {}
for ann_id in our_annotations['id'].tolist():
    position_assignments[ann_id] = random.randint(0, 3)

# Show distribution
from collections import Counter
dist = Counter(position_assignments.values())
print("Position distribution across annotations:")
for pos, (px, py) in POSITIONS.items():
    print(f"  Position {pos} (cell at x={px}, y={py} in crop): {dist[pos]} annotations")
```

```python
def compute_crop_coords(cx, cy, cell_x_in_crop, cell_y_in_crop, img_h, img_w):
    """
    Compute crop coordinates so the cell center lands at
    (cell_x_in_crop, cell_y_in_crop) within the 256x256 crop.
    Returns: x1, y1, x2, y2 in full image coordinates
             and actual cell position within crop (after clamping)
    """
    # Desired crop top-left corner
    x1 = cx - cell_x_in_crop
    y1 = cy - cell_y_in_crop
    x2 = x1 + CROP_SIZE
    y2 = y1 + CROP_SIZE

    # Clamp to image boundaries
    x1 = max(0, min(x1, img_w - CROP_SIZE))
    y1 = max(0, min(y1, img_h - CROP_SIZE))
    x2 = x1 + CROP_SIZE
    y2 = y1 + CROP_SIZE

    # Actual cell position within crop after clamping
    actual_cell_x = cx - x1
    actual_cell_y = cy - y1

    return x1, y1, x2, y2, actual_cell_x, actual_cell_y

# Test it
print("Testing crop helper:")
cx, cy = 2623, 730  # from image 007, first mitotic figure
cell_x, cell_y = POSITIONS[0]
x1, y1, x2, y2, ax, ay = compute_crop_coords(cx, cy, cell_x, cell_y, 5412, 7215)
print(f"  Cell center: ({cx}, {cy})")
print(f"  Desired cell position in crop: ({cell_x}, {cell_y})")
print(f"  Crop coords: x1={x1} y1={y1} x2={x2} y2={y2}")
print(f"  Actual cell position in crop: ({ax}, {ay})")
print(f"  Crop size: {x2-x1} x {y2-y1}")
```

```python
print("Loading SAM ViT-H...")
sam = sam_model_registry["vit_h"](checkpoint=SAM_CKPT)
predictor = SamPredictor(sam)
print("SAM loaded successfully")
```

```python
results = []
unique_image_ids = our_annotations['image_id'].unique()

for img_id in unique_image_ids:
    img_filename = images_df[images_df['id'] == img_id]['file_name'].values[0]
    img_path     = IMAGE_DIR + img_filename
    img          = np.array(Image.open(img_path).convert('RGB'))
    img_h, img_w = img.shape[:2]

    img_annotations = our_annotations[our_annotations['image_id'] == img_id]
    print(f"Processing {img_filename} — {len(img_annotations)} annotations", flush=True)

    predictor.set_image(img)

    for _, ann in img_annotations.iterrows():
        ann_id      = ann['id']
        bbox        = ann['bbox']       # [x1, y1, x2, y2] in full image
        category_id = ann['category_id']
        label       = 'mitotic' if category_id == 1 else 'hard_negative'

        # Cell center in full image
        cx = int((bbox[0] + bbox[2]) / 2)
        cy = int((bbox[1] + bbox[3]) / 2)

        # Get assigned off-center position
        pos_id               = position_assignments[ann_id]
        cell_x_in_crop, cell_y_in_crop = POSITIONS[pos_id]

        # Compute crop coordinates
        x1, y1, x2, y2, actual_cx, actual_cy = compute_crop_coords(
            cx, cy, cell_x_in_crop, cell_y_in_crop, img_h, img_w
        )

        crop = img[y1:y2, x1:x2]

        # ── Condition 1: Raw crop ──────────────────────────────────
        raw_path = f'{BASE_DIR}/raw/{label}/ann_{ann_id}.png'
        Image.fromarray(crop).save(raw_path)

        # ── Condition 2: Bounding box ──────────────────────────────
        crop_bbox = crop.copy()
        # Convert full-image bbox to crop-relative coords
        bx1 = int(max(0, bbox[0] - x1))
        by1 = int(max(0, bbox[1] - y1))
        bx2 = int(min(CROP_SIZE, bbox[2] - x1))
        by2 = int(min(CROP_SIZE, bbox[3] - y1))
        cv2.rectangle(crop_bbox, (bx1, by1), (bx2, by2), (255, 0, 0), 2)
        bbox_path = f'{BASE_DIR}/bbox/{label}/ann_{ann_id}.png'
        Image.fromarray(crop_bbox).save(bbox_path)

        # ── Condition 3: SAM mask ──────────────────────────────────
        input_box       = np.array([bbox[0], bbox[1], bbox[2], bbox[3]])
        masks, scores, _ = predictor.predict(
            point_coords=None,
            point_labels=None,
            box=input_box[None, :],
            multimask_output=False
        )
        crop_mask    = crop.copy()
        mask         = masks[0][y1:y2, x1:x2]
        colored_mask = np.zeros_like(crop_mask)
        colored_mask[mask] = [0, 255, 0]
        crop_mask    = cv2.addWeighted(crop_mask, 1.0, colored_mask, 0.5, 0)
        mask_path    = f'{BASE_DIR}/mask/{label}/ann_{ann_id}.png'
        Image.fromarray(crop_mask).save(mask_path)

        results.append({
            'ann_id'         : ann_id,
            'image_id'       : img_id,
            'filename'       : img_filename,
            'label'          : label,
            'category_id'    : category_id,
            'bbox'           : bbox,
            'crop_position'  : pos_id,
            'cell_x_in_crop' : actual_cx,
            'cell_y_in_crop' : actual_cy,
            'crop_x1'        : x1,
            'crop_y1'        : y1,
            'crop_x2'        : x2,
            'crop_y2'        : y2,
            'sam_score'      : float(scores[0]),
            'raw_path'       : raw_path,
            'bbox_path'      : bbox_path,
            'mask_path'      : mask_path,
        })

# Save annotations CSV
results_df = pd.DataFrame(results)
results_df.to_csv(f'{BASE_DIR}/annotations_exp3.csv', index=False)

print(f"\nDone!")
print(f"Total patches   : {len(results)}")
print(f"Mitotic         : {len(results_df[results_df['label']=='mitotic'])}")
print(f"Hard negative   : {len(results_df[results_df['label']=='hard_negative'])}")
print(f"Mean SAM score  : {results_df['sam_score'].mean():.3f}")
print(f"\nPosition distribution:")
print(results_df['crop_position'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt

results_df = pd.read_csv(f'{BASE_DIR}/annotations_exp3.csv')

# Pick 2 mitotic and 2 hard negative, one from each position group
samples = []
for label in ['mitotic', 'hard_negative']:
    for pos in range(4):
        subset = results_df[
            (results_df['label'] == label) &
            (results_df['crop_position'] == pos)
        ]
        if len(subset) > 0:
            samples.append(subset.iloc[0])
            break  # one per label is enough for sanity check

# Show 4 samples: 2 mitotic (pos 0, pos 1) + 2 hard_neg (pos 0, pos 1)
mitotic_samples  = results_df[results_df['label']=='mitotic'].groupby('crop_position').first().reset_index().head(2)
hardneg_samples  = results_df[results_df['label']=='hard_negative'].groupby('crop_position').first().reset_index().head(2)
show_samples     = pd.concat([mitotic_samples, hardneg_samples])

fig, axes = plt.subplots(len(show_samples), 3, figsize=(12, len(show_samples)*4))

for i, (_, row) in enumerate(show_samples.iterrows()):
    raw  = np.array(Image.open(row['raw_path']))
    bbox = np.array(Image.open(row['bbox_path']))
    mask = np.array(Image.open(row['mask_path']))

    pos_name = {0:'top-left', 1:'top-right', 2:'bottom-left', 3:'bottom-right'}

    for j, (img_arr, title) in enumerate([
        (raw,  f"{row['label']}\nPos {row['crop_position']} ({pos_name[int(row['crop_position'])]})\nRaw"),
        (bbox, f"Bounding Box\ncell at ({int(row['cell_x_in_crop'])}, {int(row['cell_y_in_crop'])})"),
        (mask, f"SAM Mask\nscore={row['sam_score']:.2f}")
    ]):
        axes[i, j].imshow(img_arr)
        axes[i, j].set_title(title, fontsize=9)
        axes[i, j].axis('off')

        # Draw dot at actual cell position on raw image
        if j == 0:
            axes[i, j].plot(row['cell_x_in_crop'], row['cell_y_in_crop'],
                           'y*', markersize=15, label='cell center')

plt.suptitle('Exp 3: Off-Center Patch Sanity Check', fontsize=13)
plt.tight_layout()
plt.show()

print(f"\nCell position stats:")
print(f"Mean cell_x in crop: {results_df['cell_x_in_crop'].mean():.1f} (expected ~128 for centered)")
print(f"Mean cell_y in crop: {results_df['cell_y_in_crop'].mean():.1f} (expected ~128 for centered)")
print(f"\nActual mean per position:")
print(results_df.groupby('crop_position')[['cell_x_in_crop','cell_y_in_crop']].mean().round(1))
```

```python
import re

# Position descriptions for the raw prompt (no bounding box or mask)
POSITION_DESCRIPTIONS = {
    0: "upper-left area",
    1: "upper-right area",
    2: "lower-left area",
    3: "lower-right area",
}

def make_prompts(pos_id):
    pos_desc = POSITION_DESCRIPTIONS[pos_id]
    return {
        'raw': f"""You are an expert pathology assistant analyzing H&E stained breast cancer tissue at 40x magnification.

A cell of interest is located in the {pos_desc} of this 256x256 pixel image.

MITOTIC FIGURES (answer YES) show:
- Condensed chromosomes in star-shaped, V-shaped, or scattered patterns
- No visible nuclear membrane (it has dissolved)
- Dark irregular chromatin arranged as distinct chromosome arms
- Chaotic, asymmetric appearance of chromosomal material

HARD NEGATIVES (answer NO) include:
- Apoptotic cells: dark shrunken nucleus, round and smooth, nuclear membrane still visible
- Hyperchromatic nuclei: very dark but round/oval shape with intact membrane
- Normal dark nuclei: uniform dark staining without chromosome structure

Be conservative. Only answer YES if you clearly see chromosome condensation patterns of active mitosis. If in doubt, answer NO.

Is the cell in the {pos_desc} of this image undergoing mitosis?
Answer in this exact format:
Yes/No, X% confidence. One sentence explaining your reasoning.
Example: "Yes, 85% confidence. The cell shows condensed chromosomal material consistent with active mitosis." """,

        'bbox': f"""You are an expert pathology assistant analyzing H&E stained breast cancer tissue at 40x magnification.

The RED BOUNDING BOX marks the specific cell to evaluate. It is located in the {pos_desc} of this image.

MITOTIC FIGURES (answer YES) show:
- Condensed chromosomes in star-shaped, V-shaped, or scattered patterns
- No visible nuclear membrane (it has dissolved)
- Dark irregular chromatin arranged as distinct chromosome arms
- Chaotic, asymmetric appearance of chromosomal material

HARD NEGATIVES (answer NO) include:
- Apoptotic cells: dark shrunken nucleus, round and smooth, nuclear membrane still visible
- Hyperchromatic nuclei: very dark but round/oval shape with intact membrane
- Normal dark nuclei: uniform dark staining without chromosome structure

Be conservative. Only answer YES if you clearly see chromosome condensation patterns of active mitosis. If in doubt, answer NO.

Is the cell inside the red bounding box undergoing mitosis?
Answer in this exact format:
Yes/No, X% confidence. One sentence explaining your reasoning.
Example: "Yes, 85% confidence. The cell shows condensed chromosomal material consistent with active mitosis." """,

        'mask': f"""You are an expert pathology assistant analyzing H&E stained breast cancer tissue at 40x magnification.

The GREEN HIGHLIGHTED REGION marks the nucleus of the specific cell to evaluate. It is located in the {pos_desc} of this image.

MITOTIC FIGURES (answer YES) show:
- Condensed chromosomes in star-shaped, V-shaped, or scattered patterns
- No visible nuclear membrane (it has dissolved)
- Dark irregular chromatin arranged as distinct chromosome arms
- Chaotic, asymmetric appearance of chromosomal material

HARD NEGATIVES (answer NO) include:
- Apoptotic cells: dark shrunken nucleus, round and smooth, nuclear membrane still visible
- Hyperchromatic nuclei: very dark but round/oval shape with intact membrane
- Normal dark nuclei: uniform dark staining without chromosome structure

Be conservative. Only answer YES if you clearly see chromosome condensation patterns of active mitosis. If in doubt, answer NO.

Is the cell in the green highlighted region undergoing mitosis?
Answer in this exact format:
Yes/No, X% confidence. One sentence explaining your reasoning.
Example: "Yes, 85% confidence. The cell shows condensed chromosomal material consistent with active mitosis." """
    }

def parse_response(text):
    """
    Parse response in format: 'Yes/No, X% confidence. Explanation.'
    Returns: (yes_no int, confidence int or None, explanation str)
    """
    text = text.strip()

    # Extract yes/no
    yes_no = -1
    if text.lower().startswith('yes'):
        yes_no = 1
    elif text.lower().startswith('no'):
        yes_no = 0

    # Extract confidence percentage
    confidence = None
    match = re.search(r'(\d+)\s*%', text)
    if match:
        confidence = int(match.group(1))

    # Extract explanation — everything after first period
    explanation = ''s
    dot_idx = text.find('.')
    if dot_idx != -1 and dot_idx < len(text) - 1:
        explanation = text[dot_idx + 1:].strip()

    return yes_no, confidence, explanation

# Test the parser
test_responses = [
    "Yes, 87% confidence. The cell shows condensed chromosomal material with no visible nuclear membrane.",
    "No, 92% confidence. The nucleus is round and dark with an intact membrane consistent with a hyperchromatic nucleus.",
    "Yes, 65% confidence. There is irregular chromatin distribution that may indicate mitosis.",
]
print("Parser test:")
for r in test_responses:
    yn, conf, expl = parse_response(r)
    print(f"  Input : {r[:60]}...")
    print(f"  yes_no={yn}, confidence={conf}%, explanation={expl[:50]}...")
    print()
```
